```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel('/content/BLUFF DETECTION MODEL.xlsx')
df
```

```python
x=df.iloc[:,1:2]
y=df.iloc[:,2]
```

```python
from sklearn.linear_model import LinearRegression
team = LinearRegression()
team.fit(x, y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='blue')
plt.plot(x,team.predict(x),color='red')
plt.title('bluff detection model(linear Regression)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
ploy = PolynomialFeatures(degree=4)
x_ploy = ploy.fit_transform(x)
linear = LinearRegression()
linear.fit(x_ploy, y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='blue')
plt.plot(x,linear.predict(x_ploy),color='red')
plt.title('bluff detection model(polynomial Regression)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
lin_pred = team.predict([[6.5]])
print(lin_pred)
```

```python
poly_pred = linear.predict(ploy.fit_transform([[6.5]]))
print(poly_pred)
```
